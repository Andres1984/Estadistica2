## Calculamos los Retornos

$ r_t = \frac{p_t}{p_{t-1}} -1$

```python
# Method 1: Using shift
retornos=pd.DataFrame()
retornos['ba'] = (ba['Close'] / ba['Close'].shift(1) ) - 1
retornos['fb'] = (fb['Close'] / fb['Close'].shift(1) ) - 1
retornos['jpm'] = (jpm['Close'] / jpm['Close'].shift(1) ) - 1
retornos['ko'] = (ko['Close'] / ko['Close'].shift(1) ) - 1

```

```python
retornos.iloc[550:844].plot(subplots=True, figsize=(9, 9))
plt.legend(loc='best')
```

## Calculamos las estadísticas descriptivas de los rendimientos

```python
a=retornos.describe()
a
```

```python
media=a.iloc[1:2]*20
media
```

```python
vol=a.iloc[2:3]*math.sqrt(20)
vol
```

```python
retornos.corr()
```

```python

sns.heatmap(retornos.corr(), annot=True,cmap="YlGnBu")
```

# Decidimos cuanto invertir en cada acción

```python
inversion = 20000
prom_reto = retornos.mean()
Pesos= np.array([0.20, 0.1, 0.1,0.6])
port_prom = prom_reto.dot(Pesos)
cov_matrix = retornos.cov()
port_vol = np.sqrt(Pesos.T.dot(cov_matrix).dot(Pesos))
prom_inv = port_prom * inversion
vol_inversion = inversion * port_vol
```

```python
port_vol*math.sqrt(20)
```

```python
prom_inv
```

```python
vol_inversion
```

```python
colors= ['blue','red','orange','green']
fig = plt.subplot(1, 1,1)
fig.scatter(vol, media, c=colors, alpha=0.3,cmap='viridis')
fig.plot(port_vol*math.sqrt(20),port_prom*20, "or")
fig.text(vol.iloc[0,0], media.iloc[0,0], "BA", fontsize=12)
fig.text(vol.iloc[0,1], media.iloc[0,1], "FB", fontsize=12)
fig.text(vol.iloc[0,2], media.iloc[0,2], "JPM", fontsize=12)
fig.text(vol.iloc[0,3], media.iloc[0,3], "KO", fontsize=12)
fig.text(port_vol*math.sqrt(20),port_prom*20, "Portafolio", fontsize=12)
fig.set_xlabel('Riesgo')
fig.set_ylabel('Rentabilidad')
```

## Medimos el riesgo de la inversión

```python
conf_level1 = 0.05
from scipy.stats import norm
corte = norm.ppf(conf_level1, prom_inv, vol_inversion)*math.sqrt(20)
corte
```

```python
var_1d1 = inversion + corte
var_1d1
```

## Realizamos el Back Testing
